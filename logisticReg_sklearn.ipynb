# Given the iris flower dataset, train a model that distinguishes the different types of iris flower, based on Logistic Regression


```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
iris_data = pd.read_csv("./data/IRIS.csv")
```

```python
iris_data
```

```python
iris_data.head()
```

```python
# now i want my X to be the features and Y to be my label
```

```python
X = iris_data.drop(columns=["species"])
Y = iris_data["species"]
```

```python
X
```

```python
Y
```

```python
# prepare model and train it
```

```python
model = LogisticRegression()
```

```python
# X.values converts it into a NumPy array (ndarray), which is the format that scikit-learn expects internally for numerical computation.
model.fit(X.values, Y)
# you can also do model.fit(X, Y) it is now accepted by sklearn
```

```python
# predict
```

```python
prediction = model.predict([[9.6,3.5,1.5,0.2]])
```

```python
print(prediction)
```
